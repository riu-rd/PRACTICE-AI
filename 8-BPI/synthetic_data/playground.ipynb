```python
import pandas as pd

df = pd.read_csv("synthetic_data.csv")
df.head(200)
```

```python
def display_case_texts(n):
    """
    Display the case_text of the first n rows in the dataframe df.
    """
    for i, text in enumerate(df['case_text'].head(n), 1):
        print(f"Case {i}:\n{text}\n{'-'*40}")

# Example usage:
# display_case_texts(5)
```

```python
display_case_texts(100)
```
